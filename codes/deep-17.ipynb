# Saving and Loading Models

```python
import numpy as np
import tensorflow.compat.v1 as tf

tf.disable_v2_behavior()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(101)
tf.set_random_seed(101)
```

## Full Network Example

Let's work on a regression example, we are trying to solve a very simple equation:

y = mx + b

y will be the y_labels and x is the x_data. We are trying to figure out the slope and the intercept for the line that best fits our data!

### Artifical Data (Some Made Up Regression Data)

```python
x_data = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
x_data
```

```python
y_label = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
plt.plot(x_data,y_label,'*')
```

** Variables **

```python
np.random.rand(2)
```

```python
m = tf.Variable(0.39,name='m')
b = tf.Variable(0.2,name='b')
```

### Cost Function

```python
error = tf.reduce_mean(y_label - (m*x_data+b))
```

### Optimizer

```python
optimizer = tf.train.GradientDescentOptimizer(
    learning_rate=0.001)
train = optimizer.minimize(error)
```

### Initialize Variables

```python

init = tf.global_variables_initializer()
```

__________
_____________
________

# Saving The Model

```python
saver = tf.train.Saver()
```

### Create Session and Run!

```python
with tf.Session() as sess:
    
    sess.run(init)
    
    epochs = 100
    
    for i in range(epochs):
        
        sess.run(train)

    # Fetch Back Results
    final_slope , final_intercept = sess.run([m,b])
    
    # ONCE YOU ARE DONE
    # GO AHEAD AND SAVE IT!
    # Make sure to provide a directory for it to make or go to. May get errors otherwise
    saver.save(sess,'new_models/my_4th_model.ckpt')
#     saver.save(sess,'new_models/my_third_model.ckpt')
```

```python
final_slope
```

```python
final_intercept
```

### Evaluate Results

```python
x_test = np.linspace(-1,11,10)
y_pred_plot = final_slope*x_test + final_intercept

plt.plot(x_test,y_pred_plot,'r')

plt.plot(x_data,y_label,'*')
```

# Loading a Model

```python
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(101)
tf.set_random_seed(101)
```

```python

m = tf.Variable(0.39,name='m')
b = tf.Variable(0.2,name='b')
init = tf.global_variables_initializer()
saver = tf.train.Saver()
    
```

```python
if tf.train.checkpoint_exists(
    tf.train.latest_checkpoint("new_models")):
    print(tf.train.latest_checkpoint("new_models"))
else:
    
    print('y')
```

```python
with tf.Session() as sess:
#     sess.run(init)
    # Restore the model
    saver.restore(sess,tf.train.latest_checkpoint("new_models"))
    

    # Fetch Back Results
    restored_slope , restored_intercept = sess.run([m,b])
```

```python
restored_slope
```

```python
restored_intercept
```

```python
x_data = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
y_label = np.linspace(0,10,10) + np.random.uniform(-1.5,1.5,10)
```

```python
x_test = np.linspace(-1,11,10)
y_pred_plot = restored_slope*x_test + restored_intercept

plt.plot(x_test,y_pred_plot,'r')

plt.plot(x_data,y_label,'*')
```
